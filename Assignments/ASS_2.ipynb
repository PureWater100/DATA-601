#### a. Read the "HW2_People.csv" with the csv module's reader function and create lists of firstnames, lastnames, birthyears, birthmonths, birthdays, weights. Note that first two lists would be strings, the following three are interegs, and the last one is a list floats.  

```
import requests
import csv

url = 'https://raw.githubusercontent.com/simsekergun/DATA601/main/2021Fall/datasets/HW2_People.csv'

with requests.Session() as s:
    download = s.get(url)

    decoded_content = download.content.decode('utf-8')

    cr = csv.reader(decoded_content.splitlines(), delimiter=',')
    next(cr) #skip header
    
    my_list = list(cr)
    for row in my_list:
        print(row)
```

```
fnames, lnames, byears, bmonths, bdays, weights = [], [], [],[], [], []
        
for row in my_list:
    fname, lname, byear = row[0], row[1], row[2] 
    bmonth, bday, weight  = row[3], row[4], row[5]  

    fnames.append(fname)
    lnames.append(lname)
    byears.append(byear)
    bmonths.append(bmonth)
    bdays.append(bday)
    weights.append(weight)

#print(fnames)
#print(lnames)
#print(byears)
#print(bmonths)
#print(bdays)
#print(weights)
```
